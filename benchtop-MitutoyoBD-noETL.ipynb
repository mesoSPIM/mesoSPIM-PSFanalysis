```python
start_time = time.time()
data = [getPSF(x, options) for x in beads]
stop_time = time.time()
print('Runtime: ',stop_time-start_time, ' s')
```

```python
centers = pd.DataFrame(centers, columns=['Z','Y','X'])
```

```python
PSF = pd.concat([x[0] for x in data])
PSF['Max'] = maxima
PSF = PSF.reset_index().drop(['index'],axis=1)
latProfile = [x[1] for x in data]
axProfile = [x[2] for x in data]
PSF = PSF.join(centers)
```

## Excitation lens field curvature
Note the curvature of the PSF detections - this is due to field curvature of the excitation lens. It also creates stripes of higher/lower axial resolution in the ETL-scanned experiment. This effect is especially strong in low-mag setting (large FOV, so excitation lens curvature is more noticeable)
