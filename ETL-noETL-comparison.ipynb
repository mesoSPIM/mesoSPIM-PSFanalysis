```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_context('paper', font_scale=2.0)
sns.set_style('ticks')
```

```python
PSF_5ms_4x_withoutETL = pd.read_csv('results/withoutETL-5ms-199hz-1x.csv')
PSF_5ms_4x_withETL = pd.read_csv('results/withETL-5ms-199hz-4x.csv')
PSF_20ms_4x_withETL = pd.read_csv('results/withETL-20ms-199hz-4x.csv')
```

```python
%matplotlib inline
fig = plt.figure(figsize=(15,10));

subfigure1 = fig.add_subplot(111)
subfigure1.plot(np.multiply(PSF_5ms_4x_withETL['X'].tolist(),1.6),PSF_5ms_4x_withETL['FWHMax'].tolist(),'.b',ms=10, label='with ETL (Freq: 199 Hz, Exp: 5 ms)')
subfigure1.plot(np.multiply(PSF_20ms_4x_withETL['X'].tolist(),1.6),PSF_20ms_4x_withETL['FWHMax'].tolist(),'.c',ms=10, label='with ETL (Freq: 199 Hz, Exp: 20 ms)')
subfigure1.plot(np.multiply(PSF_5ms_4x_withoutETL['X'].tolist(),1.6),PSF_5ms_4x_withoutETL['FWHMax'].tolist(),'.r',ms=10, label='without ETL (Freq: 199 Hz, Exp: 5 ms)')
subfigure1.set_xlim([0,3300])
subfigure1.set_ylim([0,55])
subfigure1.set_xlabel('X Distance (μm)')
subfigure1.set_ylabel('Axial FWHM (μm)')
subfigure1.legend()
plt.subplots_adjust(hspace = 0.3)
```

```python
fig.savefig('results/5ms-20ms-199Hz-4x-ETL-Comparision.svg', facecolor=fig.get_facecolor(), edgecolor='none',dpi=200)
fig.savefig('results/5ms-20ms-199Hz-4x-ETL-Comparision.jpg', facecolor=fig.get_facecolor(), edgecolor='none',dpi=200)
```
